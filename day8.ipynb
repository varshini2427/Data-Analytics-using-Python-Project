```python
import pandas as pd

data = {
    "Region": ["South", "North", "East", "West", "South", "North"],
    "Product": ["Laptop", "Mobile", "Laptop", "Tablet", "Mobile", "Tablet"],
    "Sales": [50000, 30000, 45000, 25000, 20000, 35000]
}

df = pd.DataFrame(data)

print(df)
```

```python
import pandas as pd

data = {
    "Region": ["South", "West", "North", "North", "East", "South"],
    "Product": ["Mobile", "Tablet", "Mobile", "Tablet", "Laptop", "Laptop"],
    "Sales": [20000, 25000, 30000, 35000, 45000, 50000],
}

sorted_df = pd.DataFrame(data, index=[4, 3, 1, 5, 2, 0])

print(sorted_df)

region_sales = sorted_df.groupby("Region")["Sales"].sum()
print(region_sales)
```

```python
region_sales = df.groupby("Region")["Sales"].sum()

print(region_sales)
```

```python
product_sales = df.groupby("Product")["Sales"].sum()
print(product_sales)
```

```python
result = df.groupby("Region")["Sales"].agg(["sum", "mean", "max", "min"])

print(result)
```

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [10000, 12000, 15000, 13000, 18000]

plt.plot(months, sales)
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales Amount")
plt.show()
```

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [10000, 12000, 15000, 13000, 18000]

plt.plot(
    months,
    sales,
    color="green",
    marker="o",
    linestyle="--",
)
plt.title("Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```
