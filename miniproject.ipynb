```python
df = df.drop_duplicates()

print(df)
```

```python
total_sales = df["Sales_Amount"].sum()

print("Total Sales:", total_sales)
```

```python
product_sales = df.groupby("Sales_Rep")["Sales_Amount"].sum()

print(product_sales)
```

```python
product_sales.plot(kind='bar')

plt.title("Product Sales Analysis")

plt.xlabel("Products")

plt.ylabel("Sales")

plt.show()
```
