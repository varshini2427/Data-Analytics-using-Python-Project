```python
import pandas as pd

data = {
    'emp_nm': ['varsha', 'navi'],
    'emp_sal': [25000, 30000]
}

df = pd.DataFrame(data)

df.rename(columns={
    'emp_nm': 'Employee Name',
    'emp_sal': 'Employee Salary'
}, inplace=True)

print(df)
```
