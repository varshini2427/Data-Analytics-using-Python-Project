```python
scores = [58, 45, 73, 89, 99]

total = sum(scores)
average = total / len(scores)
highest = max(scores)
lowest = min(scores)

print("students scores:", scores)
print("Total marks:", total)
print("average marks:", average)
print("Highest marks:", highest)
print("lowest marks:", lowest)
```
